## Preview
The tweets are taken from July 2015 to December 2015. The tweets are in Indonesian and extracted by using 2 keywords : 'avanza' and '@Toyota_ID'. Snscrape module is used to extract the tweets.

Based on my personal experience, the number of tweets when I extracted the tweets all at once were not as many as the tweets when I separated the timeframe monthly. Hence, you could see the codes are repeating here.

```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

#### Extracting tweet from July 2015

Keyword : 'avanza'

```python
#July 2015
# Setting variables to be used below
maxTweets = 1000000000

# Creating list to append tweet data to
tweets_july2015 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('avanza lang:id since:2015-07-01 until:2015-07-31').get_items()):
    if i>maxTweets:
        break
    tweets_july2015.append([tweet.date, tweet.id, tweet.content, tweet.user.username])
```

```python
# Creating a dataframe from the tweets list above
tweets_july2015 = pd.DataFrame(tweets_july2015, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])

# Display dataframe
tweets_july2015
```

```python
# Remove timezone from columns
tweets_july2015['Datetime'] = tweets_july2015['Datetime'].dt.tz_localize(None)
```

Keyword : @Toyota_ID

```python
#July 2015
# Setting variables to be used below
maxTweets = 1000000000

# Creating list to append tweet data to
tweets_july2015_toyota = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('@ToyotaID lang:id since:2015-07-01 until:2015-07-31').get_items()):
    if i>maxTweets:
        break
    tweets_july2015_toyota.append([tweet.date, tweet.id, tweet.content, tweet.user.username])
```

```python
# Creating a dataframe from the tweets list above
tweets_july2015_toyota = pd.DataFrame(tweets_july2015_toyota, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])

# Display dataframe
tweets_july2015_toyota
```

```python
# Remove timezone from columns
tweets_july2015_toyota['Datetime'] = tweets_july2015_toyota['Datetime'].dt.tz_localize(None)
```

```python
#append 2 dataframes together
result = tweets_july2015.append(tweets_july2015_toyota)
display(result)
result.to_excel('July 2015.xlsx')
```
